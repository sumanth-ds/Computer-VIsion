```python
import numpy as np
import cv2
import pandas as pd
import matplotlib.pyplot as plt


def viewImage(image, name_of_window):
    cv2.namedWindow(name_of_window, cv2.WINDOW_NORMAL)
    cv2.imshow(name_of_window, image)
    cv2.waitKey()
    cv2.destroyAllWindows()
```

```python
img = cv2.imread('Heroine.jpg', 0)
viewImage(img, 'Heroine')
```

```python
img = cv2.imread('Heroine.jpg', 1)
viewImage(img, 'Heroine')
```

```python
scale_percent = 30
width = int(img.shape[1] * scale_percent/100)
height = int(img.shape[1] * scale_percent/100)
dim = (width, height)

resized_img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
viewImage(resized_img , "After resizing the img with 30%")
```

# Image Rotating

```python
img.shape
```

```python
(h, w, d) = img.shape
center = (w/2, h/2)
m = cv2.getRotationMatrix2D(center, 24, 1.0)
rotated = cv2.warpAffine(img, m, (w,h))
viewImage(rotated, 'image after rotation')
```

# Threshold

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
ret, threshold_image = cv2.threshold(gray_image, 50, 255, 0)
viewImage(threshold_image, 'heroine')
```

# Object Detection

```python

face_classifier = cv2.CascadeClassifier("haarcascades/haarcascade_frontalface_default.xml")
image = cv2.imread('Heroine.jpg')
faces = face_classifier.detectMultiScale(image, 1.3 , 5)
if faces is ():
    print('No Faces Found')
for (x,y,w,h) in faces:
    cv2.rectangle(image, (x,y), (x+w, y+h), (0,0,255), 2)
    cv2.imshow('Face Detection', image)
    cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
face_classifier = cv2.CascadeClassifier("haarcascades/haarcascade_frontalface_default.xml")
face_classifier = cv2.CascadeClassifier("haarcascades/haarcascade_frontalface_default.xml")
image = cv2.imread('multiple_faces.jpg')
faces = face_classifier.detectMultiScale(image, 1.3 , 5)
if faces is ():
    print('No Faces Found')
for (x,y,w,h) in faces:
    cv2.rectangle(image, (x,y), (x+w, y+h), (0,0,255), 2)
    cv2.imshow('Face Detection', image)
    cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
import pandas as pd
import matplotlib.pyplot as plt

face_classifier = cv2.CascadeClassifier("haarcascades/haarcascade_frontalface_default.xml")
eye_classifier = cv2.CascadeClassifier("haarcascades/haarcascade_eye.xml")
image = cv2.imread('multiple_faces.jpg')
faces = face_classifier.detectMultiScale(image, 1.3 , 5)
if faces is ():
    print('No Faces Found')
for (x,y,w,h) in faces:
    cv2.rectangle(image, (x,y), (x+w, y+h), (0,0,255), 2)
    cv2.imshow('Face Detection', image)
    cv2.waitKey(0)
    eye_image = image[y:y+h, x:x+w]
    eyes = eye_classifier.detectMultiScale(eye_image)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(eye_image, (ex,ey), (ex+ew, ey+eh), (255,0,0), 2)
        cv2.imshow('eye', image)
        cv2.waitKey(0)


cv2.destroyAllWindows()
```
