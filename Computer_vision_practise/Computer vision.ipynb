# Multiple color at one 

```python
color_list = [method for method in dir(cv2) if method.startswith('COLOR_') is True]
color_list
```

```python
len(color_list)
```

```python
input = cv2.imread('dog.JPG')
for color in range(1,len(color_list)):
    color_name = color_list[color]
    colr_code = 'cv2.'+color_name
    resize = cv2.resize(input, (400,400))
    try:
        reg_image = cv2.cvtColor(resize, eval(colr_code))
        cv2.imshow(col_name+"_"+str(color), reg_image)
        cv2.waitKey()
    except Exception as e:
        print('Error on',str(colr_code),e)
    cv2.destroyAllWindows()
```

# Polygone

```python
black = np.zeros((512,512, 3), np.uint8)
pts = np.array([[10,15], [400,50],[90,200], [50,500]], np.int64)
pts = pts.reshape(-1,1,2)
cv2.polylines(black, [pts], True, (255,255,0), 3)
cv2.imshow("polygon", black)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Sharpening

```python
image = cv2.imread('dog1.jpeg')
cv2.imshow('Original image', image)

sharpen_array = np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])

sharpening = cv2.filter2D(image,-1, sharpen_array )
cv2.imshow('sharpening',sharpening )

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Thresholding , Binarization and adaptive thresholding

```python
image = cv2.imread('dog1.jpeg',0)
cv2.imshow('image', image)
ret, thresh1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
cv2.imshow('Threshold image', thresh1)

ret, thresh2 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)
cv2.imshow('Threshold image', thresh2)



cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2

image = cv2.imread('dog1.jpeg',0)
image1 = cv2.GaussianBlur(image, (3,3), 0)
thresh3 = cv2.adaptiveThreshold(image1,  255, cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY ,3,5)
cv2.imshow('Threshold image', thresh3)
cv2.waitKey(0)
cv2.destroyAllWindows()

ret, thresh2 = cv2.threshold(image1, 0, 255, cv2.THRESH_BINARY+ cv2.THRESH_OTSU)
cv2.imshow('Threshold image', thresh2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# EDGE Detection

```python

image = cv2.imread('dog1.jpeg',0)
cv2.imshow('Original image', image)


height, width = image.shape

# Extrac Sobel EDges
Sobel_x = cv2.Sobel(image,cv2.CV_64F,0,1, ksize =5)
Sobel_y = cv2.Sobel(image,cv2.CV_64F,1,0, ksize =5)

cv2.imshow('Sobel_x image', Sobel_x)

cv2.imshow('Sobel_y image', Sobel_y)



# Laplactin

laplacian = cv2.Laplacian(image,cv2.CV_64F)

cv2.imshow('laplacian image', laplacian)



# canny - is very famous and it will remove noise whiite detecting edge

canny = cv2.Canny(image, 50, 120)
cv2.imshow('Canny' , canny)


cv2.waitKey(0)

cv2.destroyAllWindows()
```

# Live Project

```python
import numpy as np
import cv2


def sketch(image):
    # convert image to grayscale
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Blur the image
    img_gray_blur = cv2.GaussianBlur(img_gray, (5,5),0)
    # extract edges
    canny_edges = cv2.Canny(img_gray_blur, 10, 70)
    # Threshold
    ret, mask = cv2.threshold(canny_edges, 70, 255, cv2.THRESH_BINARY_INV)
    return mask

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    cv2.imshow('Show me live sketch', sketch(frame))
    if cv2.waitKey(1) == 13:
        break
cap.release()
cv2.destroyAllWindows()
```
